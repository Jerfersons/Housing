```python
result=df.groupby('location')['Price (in rupees)'].mean()
```

```python
fig, ax = plt.subplots(figsize=(28,12))

ax.bar(result.index, result.values)

ax.set_ylabel('maximum price of houses')
ax.set_xlabel('locations')
ax.set_title('maximum price of houses')
plt.xticks(rotation = 90) # Rotates X-Axis Ticks by 45-degrees

plt.show()
```

```python
df['sqft area']=df['Carpet Area'].str.replace('\D', '', regex=True)
df['sqft area']= df['sqft area'].fillna(0)
df['sqft area']=df['sqft area'].astype('int')
sqft_area=df.groupby('location')['sqft area'].mean()
```

```python
fig, ax = plt.subplots(figsize=(28,12))
ax.bar(sqft_area.index, sqft_area.values)

ax.set_ylabel('sqft area')
ax.set_xlabel('locations')
plt.xticks(rotation = 90) # Rotates X-Axis Ticks by 45-degrees
ax.set_title('Mean sqft area by location')

plt.show()
```

Colum to fill empty values with Zero

```python
df['Status_numeric']= df['Status'].fillna(0)
df['Status_numeric']= df[df['Status_numeric']=='Ready to Move']=1

df.info()
```

Plotting the graph of location by status
    - The status column contains apartments that are ready to move and not ready to move
    

```python
location_status=df.groupby('location')['Status_numeric'].sum()
```

```python
fig, ax = plt.subplots(figsize=(28,12))
ax.bar(location_status.index, location_status.values)
ax.set_ylabel('number of houses that are ready to move')
ax.set_xlabel('locations')
ax.set_title('Distribution of houses that are ready to be moved into.')


plt.xticks(rotation = 90) # Rotates X-Axis Ticks by 45-degrees


plt.show()
```

Based  on the data, New Dehli has the highest amount of houses to be moved into due to having the highest total number of houses in the data set.
